```python
import matplotlib.pyplot as plt
import numpy as np

def evaluate_and_visualize(model_dict):
    num_models = 4
    num_images = 5

    # Create a subplot grid for displaying images and predictions
    fig, axs = plt.subplots(num_images, num_models + 1, figsize=(4 * (num_models + 1), 4 * num_images))

    # Iterate over models and names to display column headlines
    for i, (model_name, model) in enumerate(model_dict.items()):
        axs[0, i].set_title(model_name)

    # Display "Ground truth" heading in the last column
    axs[0, num_models].set_title("Ground truth")

    eval_dict = {
        "model1": {
            "ap": [0.5, 0.6, 0.7, 0.8, 0.9, 0.95],
            "ar": [0.5, 0.6, 0.7, 0.8, 0.9, 0.95]
        },
        "model2": {
            "ap": [0.5, 0.6, 0.7, 0.8, 0.9, 0.95],
            "ar": [0.5, 0.6, 0.7, 0.8, 0.9, 0.95]
        },
        "model3": {
            "ap": [0.5, 0.6, 0.7, 0.8, 0.9, 0.95],
            "ar": [0.5, 0.6, 0.7, 0.8, 0.9, 0.95]
        },
        "model4": {
            "ap": [0.5, 0.6, 0.7, 0.8, 0.9, 0.95],
            "ar": [0.5, 0.6, 0.7, 0.8, 0.9, 0.95]
        }
    }

    plt.tight_layout()
    

    # Extract the model names and evaluation metrics from eval_dict
    model_names = list(eval_dict.keys())
    ap_values = [eval_dict[model_name]['ap'] for model_name in model_names]
    ar_values = [eval_dict[model_name]['ar'] for model_name in model_names]

    # Create the bar chart
    fig, axs = plt.subplots(num_images, 1, figsize=(4 * (num_models + 1), 4 * num_images))

    # Set the positions and width for the bars
    bar_width = 0.35
    index = np.arange(len(model_names))

    # Plot the AP bars
    for i in range(num_images):
        ap = ap_values[i]
        axs[i].bar(index, ap, bar_width, label='AP')
        axs[i].set_ylabel('AP')
        axs[i].set_xticks(index)
        axs[i].set_xticklabels(model_names)
        axs[i].legend()

    # Plot the AR bars
    for i in range(num_images):
        ar = ar_values[i]
        axs[i].bar(index + bar_width, ar, bar_width, label='AR')
        axs[i].set_ylabel('AR')
        axs[i].set_xticks(index)
        axs[i].set_xticklabels(model_names)
        axs[i].legend()

    # Adjust the layout and display the chart
    plt.tight_layout()
    plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_eval_metrics(eval_dict):
    models = list(eval_dict.keys())
    metrics = list(eval_dict[models[0]].keys())
    num_models = len(models)
    num_metrics = len(metrics)

    fig, axs = plt.subplots(1, num_models, figsize=(15, 5))

    for i, model in enumerate(models):
        ap_values = eval_dict[model]["ap"]
        ar_values = eval_dict[model]["ar"]

        x = np.arange(len(ap_values))
        width = 0.35

        axs[i].bar(x - width/2, ap_values, width, label='AP')
        axs[i].bar(x + width/2, ar_values, width, label='AR')

        axs[i].set_ylabel('Value')
        axs[i].set_title(model)
        axs[i].set_xticks(x)
        axs[i].set_xticklabels([''] * len(ap_values))  # Clear x-axis labels

        axs[i].legend()

        # Add value labels to the bars
        for j, ap_val in enumerate(ap_values):
            axs[i].annotate(f'{ap_val:.2f}', xy=(x[j] - width/2, ap_val), xytext=(0, 3),
                            textcoords="offset points", ha='center', va='bottom')
        for j, ar_val in enumerate(ar_values):
            axs[i].annotate(f'{ar_val:.2f}', xy=(x[j] + width/2, ar_val), xytext=(0, 3),
                            textcoords="offset points", ha='center', va='bottom')

        # Rotate and align the value labels
        axs[i].tick_params(axis='y', labelrotation=90, pad=10, length=0)

    # Add shared x-axis label
    fig.text(0.5, 0.01, 'Threshold', ha='center')

    plt.tight_layout()
    plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

eval_dict = {
    "model1": {
        "ap": [0.5, 0.6, 0.7, 0.8, 0.9, 0.95],
        "ar": [0.5, 0.6, 0.7, 0.8, 0.9, 0.95]
    },
    "model2": {
        "ap": [0.5, 0.6, 0.7, 0.8, 0.9, 0.95],
        "ar": [0.5, 0.6, 0.7, 0.8, 0.9, 0.95]
    },
    "model3": {
        "ap": [0.5, 0.6, 0.7, 0.8, 0.9, 0.95],
        "ar": [0.5, 0.6, 0.7, 0.8, 0.9, 0.95]
    },
    "model4": {
        "ap": [0.5, 0.6, 0.7, 0.8, 0.9, 0.95],
        "ar": [0.5, 0.6, 0.7, 0.8, 0.9, 0.1]
    }
}

labels_ap = [
    "0.50:0.95_all",
    "0.50_all",
    "0.75_all",
    "0.50:0.95_small",
    "0.50:0.95_medium",
    "0.50:0.95_large"
]

labels_ar = [
    "0.50:0.95_all",
    "0.50:0.95_all",
    "0.50:0.95_all",
    "0.50:0.95_small",
    "0.50:0.95_medium",
    "0.50:0.95_large"
]

models = eval_dict.keys()
ap_values = [eval_dict[model]["ap"] for model in models]
ar_values = [eval_dict[model]["ar"] for model in models]

num_models = len(models)
num_values = len(labels_ap)

fig, axes = plt.subplots(num_models, 2, figsize=(12, num_models * 4))
plt.subplots_adjust(hspace=1,)

for i, model in enumerate(models):
    axes[i, 0].bar(np.arange(num_values), ap_values[i], align='center')
    axes[i, 0].set_xticks(np.arange(num_values))
    axes[i, 0].set_xticklabels(labels_ap, rotation=45)
    axes[i, 0].set_ylabel("AP")
    axes[i, 0].set_title(f"Model {model} - AP")

    axes[i, 1].bar(np.arange(num_values), ar_values[i], align='center')
    axes[i, 1].set_xticks(np.arange(num_values))
    axes[i, 1].set_xticklabels(labels_ar, rotation=45)
    axes[i, 1].set_ylabel("AR")
    axes[i, 1].set_title(f"Model {model} - AR")

plt.show()
```
